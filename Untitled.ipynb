```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import statsmodels.api as sm
import pystan
```

```python
def create_lag(df, var_names):
    
    df_out = df.copy()
    
    for key in var_names.keys():
        for i in range(1, var_names[key] + 1):
            
            df_out[key + '_tm' + str(i)] = df[key].shift(i)
            
    return df_out.dropna()
            
            
```

```python
# Simulated data
np.random.seed(1234)

nobs = 101

x1 = np.random.normal(size=nobs)
x2 = np.random.normal(size=nobs)
x3 = np.random.normal(size=nobs)
x4 = np.random.normal(size=nobs)


# X = np.vstack([np.ones(nobs), x1, x2, x3]).T
X = np.vstack([x1, x2, x3, x4]).T
```

```python
def get_prior_type(var_names):
    
    """
    Description: This function recieves a list of string containing the names of the variables in a model
    for current account and assing a prior type for each of them. Here, we allow for two type, 1 and 2,
    which will be translated to a prior defined only for negative values and a prior defined only for 
    positive value, respectively. Since we know from Economic Theory that REER and GDP is negative related
    to CA they recieve a type 1 prior. Since TOT is positive related with CA, it recieves a type 2 prior.
    """
    
    var_names = [x[4:] for x in var_names]    
    mapping = {"REER": 1, "GDP": 1, "TOT": 2}

    result = []
    for s in var_names:
        for prefix, number in mapping.items():
            if s.startswith(prefix):
                result.append(number)

            
    n_type1 = sum(np.array(results)==1)
    n_type2 = sum(np.array(results)==2)

    return result, n_type1, n_type2
```

```python
results, n_type1, n_type2 = get_prior_type(var_names)
```

```python
results, n_type1, n_type2
```

```python
def inital_values():
    beta1 = np.random.uniform(low=-1, high=0, size=n_type1+1)
    beta2 = np.random.uniform(low=0, high=1, size=n_type2)
    sigma = np.random.gamma(shape=0.25, scale=4, size=1)

    dict_init = {'beta': list(np.hstack([beta1, beta2])), 'sigma': float(sigma)}
    
    return dict_init
```

```python
X = sm.add_constant(df_reg[var_names]).values


sigma2 = 2
beta = np.array([5, -9, -5, -1, -4, -5, -5, 5, -5])

y = X.dot(beta) + np.random.normal(size=nobs-1, scale=sigma2)
```

```python
model= '''data {
    int<lower=0> N;   // number of data items
    int<lower=0> K;   // number of predictors
    matrix[N, K] x;   // predictor matrix
    vector[N] y;      // outcome vector
    int<lower=0> n_type1;
    int<lower=0> n_type2;
}

parameters {
    vector[K] beta;       // coefficients for predictors
    real<lower=0> sigma;  // error scale
}

model {
    
    beta[1] ~ normal(0,10);
    
    // type 1 prior
    for (i in 2:n_type1){
        beta[i]~ uniform(-1000,0); // restrict beta to be negative
    }
    
    // type 2 prior
    for (i in (n_type1+2):(n_type1 + n_type2+1)){
        beta[i] ~ uniform(0,1000); // restrict beta to be positive
    }
    
    sigma ~ gamma(0.25, 0.25); 
    y ~ normal(x * beta , sigma);  // likelihood
}'''
```

```python
reg = pystan.StanModel(model_code=model) # compila o modelo
```

```python
data = {"N": len(y),
        "K": X.shape[1],
        "n_type1": n_type1,
        "n_type2": n_type2,
        "y": y,
        "x": X}


fit = reg.sampling(data=data, init=inital_values, iter=2000, warmup= 1000, chains=2, seed=1, verbose=True,
                 control = {'adapt_delta': .8}) # roda as iterações
```

```python
summary = pd.DataFrame(fit.extract()['beta'], columns=['const']+var_names).describe().T
```

```python
summary['count'] = beta
summary['OLS'] = sm.OLS(y, X).fit().params
summary.rename(columns={'count': 'beta'}, inplace=True)
summary
```
